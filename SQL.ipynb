## conn + cursor way
Python-native control (more production-like)

```python
# Setup Connection
```

```python
import sqlite3

conn = sqlite3.connect("students.db")
cursor = conn.cursor()
```

Data Exploration

```python
# View tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

View students

```python
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

View specific columns

```python
cursor.execute("SELECT name, score FROM students")
print(cursor.fetchall())
```

Filtering

```python
cursor.execute("""
SELECT *
FROM students
WHERE score > 85
""")

cursor.fetchall()
```

Sorting

```python
cursor.execute("""
SELECT *
FROM students
ORDER BY score DESC
""")

print(cursor.fetchall())
```

Aggregations

```python
# Average score
cursor.execute("""
SELECT AVG(score)
FROM students
""")

print(cursor.fetchone())
```

Group by course

```python
cursor.execute("""
SELECT course, AVG(score)
FROM students
GROUP BY course
""")

cursor.fetchall()
```

Insert data

```python
cursor.execute("""
INSERT INTO students (name, age, course, score)
VALUES (?, ?, ?, ?)
""", ("Faith", 23, "AI", 87.4))

conn.commit()
```

```python
cursor.execute("""
SELECT *
FROM students
""")

cursor.fetchall()
```

Update data

```python
cursor.execute("""
UPDATE students
SET score = 95
WHERE name = ?
""", ("Noah",)
)

conn.commit()
```

```python
cursor.execute("""
SELECT *
FROM students
""")

cursor.fetchall()
```

Delete

```python
cursor.execute("""
DELETE FROM students
WHERE name = ?
""", ("Faith",))

conn.commit()
```

```python
cursor.execute("""
SELECT *
FROM students
""")

cursor.fetchall()
```

Create new table

```python
cursor.execute("""
CREATE TABLE instructors (
    instructor_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT
)
""")

conn.commit()
```

Insert Instructors

```python
instructors_data = [
    (1, "Dr. Kimani", "Data Science"),
    (2, "Prof. Achieng", "Statistics"),
    (3, "Dr. Mwangi", "AI")
]

cursor.executemany("""
INSERT INTO instructors (instructor_id, name, department)
VALUES (?, ?, ?)
""", instructors_data)

conn.commit()
```

Add instructors to the students

```python
cursor.execute("""
ALTER TABLE students
ADD COLUMN instructor_id INTEGER
""")

conn.commit()
```

```python
cursor.execute("""
UPDATE students SET instructor_id = 1 WHERE course = 'Data Science'
""")

cursor.execute("""
UPDATE students SET instructor_id = 2 WHERE course = 'Statistics'
""")

cursor.execute("""
UPDATE students SET instructor_id = 3 WHERE course = 'AI'
""")

conn.commit()
```

Join tables

```python
cursor.execute("""
SELECT s.name, s.course, i.name
FROM students s
JOIN instructors i
ON s.instructor_id = i.instructor_id
""")

print(cursor.fetchall())
```

Convert to pandas

```python
import pandas as pd

cursor.execute("SELECT * FROM students")
data = cursor.fetchall()

df = pd.DataFrame(data, columns=["id", "name", "age", "course", "score"])
df.head()
```

Close Connection (Good practice)

```python
conn.close()
```
